```python
import yfinance as yf
# Set the start and end dates
start_date = "2015-01-01"
end_date = "2024-10-31"

# Download data for TSLA, BND, and SPY
tsla_data = yf.download("TSLA", start=start_date, end=end_date)
bnd_data = yf.download("BND", start=start_date, end=end_date)
spy_data = yf.download("SPY", start=start_date, end=end_date)
```

```python
# Save data to CSV
tsla_data.to_csv("../data/raw/tsla_data.csv")
bnd_data.to_csv("../data/raw/bnd_data.csv")
spy_data.to_csv("../data/raw/spy_data.csv")
```

```python
data_tsla = pd.read_csv("../data/raw/tsla_data.csv")
data_tsla.head()
```

```python
# Reset column names correctly

data_tsla.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
data_tsla = data_tsla.drop([0, 1])
# Convert 'Date' to datetime
data_tsla['Date'] = pd.to_datetime(data_tsla['Date'], errors='coerce')

# Save cleaned data
data_tsla.to_csv("../data/processed/cleaned_tsla_data.csv", index=False)
```
